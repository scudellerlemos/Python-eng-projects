```python
#Com o tratamento correto, vamos começar a tradução da coluna Pais_Origem

df_translated_all = df_translated.replace({"Pais_Origem":pais_lang})
df_translated_all.head()
```

```python
#Agora vamos criar uma metrica unica para avaliação das series baseada nos indicadores que possuimos, iremos utilizar a equação abaixo:

# Weighted Rating (WR) = ((v*(v+m))*r) + ((m*(v+m))*c)

#The following apply:

#v is the number of votes garnered by the movie
#m is the minimum number of votes required for the movie to be in the chart (the prerequisite)
#R is the mean rating of the movie
#C is the mean rating of all the movies in the dataset

m = df_translated_all["Contagem_Votos"].quantile(0.80)
C = df_translated_all["Nota_Votos"].mean()

def weighted_rating(x, m=m, C=C):
    v = x['Contagem_Votos']
    R = x['Nota_Votos']
    ## Compute the weighted score
    return (v/(v+m) * R) + (m/(m+v) * C)


df_translated_all["Score"] = df_translated_all.apply(weighted_rating, axis=1)

df_translated_all.head()
```

```python
#Filtrando TV shows e Series com Score >8
df_translated_all_query = df_translated_all.query('Score > 8')

#Agrupando os campos escolhidos
df_translated_all_query_group = df_translated_all_query.groupby(["ID","Nome","Nome_Origem","Idioma_Origem","Pais_Origem"]).agg({"Popularidade":"sum","Nota_Votos":"sum","Contagem_Votos":"sum"})
df_translated_all_query_group = df_translated_all_query_group.reset_index()
df_translated_all_query_group.set_index("ID",inplace=True)
```
