```python
import pandas as pd
filesaved = "datafolder/netflix_data/NETFLIX_TABLE.csv"
filepath = "datafolder/netflix_data/titles.csv"
```

```python
#lendo arquivos e construindo dataframe
df_netflix = pd.read_csv(filepath)
df_netflix.head()
```

```python
#puxando informações das colunas
df_netflix.info()
```

```python
#dropando colunas que não iremos utilizar
df_netflix.drop(["age_certification","seasons","imdb_id","imdb_score","imdb_votes","tmdb_popularity","tmdb_score"],axis= 1,inplace=True)
df_netflix.head()
```

```python
#iniciando tratamento de dados
#retirando as linhas em formato de linhas e colocando no formato desejato

df_netflix_tratada=df_netflix

#Tratando quando a lista é vazia com brakets "[]"
df_netflix_tratada["genres"].replace("[]","NaN",inplace=True)
df_netflix_tratada["production_countries"].replace("[]","NaN",inplace=True)

#Tratando para retirar os brakets "[" e "]" da lista das colunas genres e countries
df_netflix_tratada["genres"]= df_netflix_tratada["genres"].str.split("[")
df_netflix_tratada= df_netflix_tratada.explode("genres")
df_netflix_tratada["genres"]= df_netflix_tratada["genres"].str.split("]")
df_netflix_tratada= df_netflix_tratada.explode("genres")

df_netflix_tratada["production_countries"]= df_netflix_tratada["production_countries"].str.split("[")
df_netflix_tratada= df_netflix_tratada.explode("production_countries")
df_netflix_tratada["production_countries"]= df_netflix_tratada["production_countries"].str.split("]")
df_netflix_tratada= df_netflix_tratada.explode("production_countries")

#Tratando as strings ", " da coluna genres
df_netflix_tratada["genres"].replace(regex=", ",value="/",inplace=True)
df_netflix_tratada["genres"]= df_netflix_tratada["genres"].str.split("/")
df_netflix_tratada= df_netflix_tratada.explode("genres")

#Trocando as strings vazias por númerios vazios
df_netflix_tratada["production_countries"].replace(regex="",value=float("NaN"),inplace=True)
df_netflix_tratada["genres"].replace(regex="",value=float("NaN"),inplace=True)

#Dropando linhas vazias
df_netflix_tratada.dropna(subset=["genres","production_countries"],inplace=True)
```

```python
#regeitando a string " ' " nas colunas genres e production_countries
df_netflix_tratada.replace(regex="'",value="",inplace=True)
df_netflix_tratada.head()
```

```python
#agrupando multiplos paises dentro da coluna production_countries
country_condition= df_netflix_tratada["production_countries"].apply(len)> 2

df_netflix_tratada["production_countries"].loc[country_condition == True]="MULTIPLE COUNTRIES"


```

```python
#Realizando filtros atraves do metodo .query
df_netflix_query=df_netflix_tratada.query('runtime > 60 and type == "SHOW"  and release_year > 2000')


#agrupando campos finais e nas agregações desejadas
df_netflix_groupby=df_netflix_query.groupby(["genres","production_countries"]).agg({"runtime":"mean","title":"count"})
df_netflix_groupby=df_netflix_groupby.reset_index()
df_netflix_groupby.head()
```

```python
df_netflix_groupby.to_csv(filesaved)
```
